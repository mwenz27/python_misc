## Decision Trees

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import *
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split 
# X, y = load_iris(return_X_y=True)
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

iris = load_iris()
x = iris['data']
y = iris['target']
```

```python
y
# Class [ 0, 1, 2]
# ['setosa', 'versicolor', 'virginica']
```

```python
iris.target_names
```

```python
print(len(x))
print(len(y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=40)
# try different random states
```

```python
clf = tree.DecisionTreeClassifier() # clf is like a classifier variable, sometimes called a model
```

```python
clf =  clf.fit(X_train, y_train)
clf.predict(X_test)
```

```python
y_test
```

```python
# X_test[:] - y_test[:]
```

```python
fig = plt.figure(figsize=(10,10))
tree.plot_tree(clf, filled=True, class_names=iris.target_names, feature_names=iris.feature_names)
```

## Random Forrest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators=5, max_depth=5, random_state=0)
clf.fit(X_train, y_train)
```

```python
print(clf.feature_importances_)
```
