## Introduction to Matplot lib, numpy and pandas

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot([1,2,3,4])
plt.ylabel('my label')
plt.show()
plt.plot([1,2,3,4], [2,4,6,8], color='r', marker= 'o')
plt.axis([0, 5, 0, 10])
plt.show()
x = np.linspace(-5, 5, 50)
y = np.sin(x)     
plt.plot(x, y, marker="x") 
plt.show()
```

```python
t = np.arange(0., 5., 0.2) 
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^') 
plt.show() 


names = ['group_a', 'group_b', 'group_c'] 
values = [1, 10, 100] 
plt.bar(names, values) 
plt.show()
```
