```python
def find_header(file):
    with open(file, 'r') as fir:
        file_lines = fir.readlines()
    header = 0
    for key, index in enumerate(file_lines):
        if "NLAYR\t 20\n" in index.rstrip():
            header = key + 2
        return header # this is added as the text files has 2 more lines after the finding header
```
